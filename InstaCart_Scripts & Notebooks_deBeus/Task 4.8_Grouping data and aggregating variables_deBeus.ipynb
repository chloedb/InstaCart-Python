### Step Six - Spending column

### Step Seven - Frequency column

```python
#first let's create the variable we will group by and see what it typically looks like across users
all_prior_orders.groupby('user_id').agg({'days_since_prior_order': ['median']})
```

```python
all_prior_orders['order_frequency_median'] = all_prior_orders.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
all_prior_orders.head()
```

```python
#now let's distribute the variable of order_frequency_median to decide the freqency of the customer
all_prior_orders.loc[all_prior_orders['order_frequency_median'] > 20, 'frequency_status'] = 'Non-frequent customer'
```

```python
all_prior_orders.loc[(all_prior_orders['order_frequency_median'] > 10) & (all_prior_orders['order_frequency_median'] <= 20), 'frequency_status'] = 'Regular frequency customer'
```

```python
all_prior_orders.loc[all_prior_orders['order_frequency_median'] <= 10, 'frequency_status'] = 'Frequent customer'
```

```python
all_prior_orders['frequency_status'].value_counts(dropna = False)
```
