### If statements

```python
#using def to define the variable that you are calling when you run this
def price_label(row):
    if row['prices'] <= 5:
        return 'Low-range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid-range product'
    elif row['prices'] > 15:
        return 'High range product'
    else: return 'Not enough data'
```

```python
prior_subset['price_range'] = prior_subset.apply(price_label, axis = 1)
prior_subset['price_range'].value_counts(dropna = False)
```

```python
#there are no high range or 'no data', so let's look
prior_subset['prices'].max()
```

### Using loc() instead of if statement

```python
prior_subset.loc[prior_subset['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
prior_subset.loc[(prior_subset['prices'] <= 15) & (prior_subset['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
prior_subset.loc[prior_subset['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

### If-statements with For-Loops

```python
all_prior_orders['orders_day_of_week'].value_counts(dropna = False)
```

```python
# saturday = 0 is the busiest; wednesday = 4 is the slowest
result = []

for value in all_prior_orders['orders_day_of_week']:
    if value == 0:
        result.append('Busiest day')
    elif value == 4:
        result.append('Least busy')
    else:
        result.append('Regularly busy')
```

```python
result
```

```python
#create a new column based on this variable we created/defined
all_prior_orders['busiest_days'] = result
all_prior_orders['busiest_days'].value_counts(dropna = False)
```

```python
# create price_label column
all_prior_orders['price_category'] = all_prior_orders['price_range_loc']
all_prior_orders['price_category'].value_counts(dropna = False)
```
