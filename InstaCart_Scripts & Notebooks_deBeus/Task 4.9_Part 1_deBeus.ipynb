#### Step 6 - combine data

```python
#first, pull in the other data
all_prior_orders = pd.read_pickle(os.path.join(mainPath, '02 Datasets', 'Clean data', 'orders_products_merged.pkl'))
#there's a memory error, so let's try to merge in the other notebook
```

```python
print(all_prior_orders.dtypes)
print(customers.dtypes)
```

```python
all_prior_orders['new_customer']=all_prior_orders['new_customer'].astype('int32')
all_prior_orders['days_since_prior_order']=all_prior_orders['days_since_prior_order'].astype('int8')
all_prior_orders['aisle_id']=all_prior_orders['aisle_id'].astype('int8')
all_prior_orders['department_id']=all_prior_orders['department_id'].astype('int8')
```

```python
customers['user_id']=customers['user_id'].astype('int8')
customers['age']=customers['age'].astype('int8')
customers['income']=customers['income'].astype('int32')
customers['dependents']=customers['dependents'].astype('int8')
```

```python
all_prior_orders.drop('Unnamed: 0_y', axis=1, inplace=True)
all_prior_orders.drop('time_period_loc', axis=1, inplace=True)
all_prior_orders.drop('price_range_loc', axis=1, inplace=True)
```

```python
print(all_prior_orders.dtypes)
print(customers.dtypes)
```

```python
all_prior_orders['spending_mean']=all_prior_orders['spending_mean'].astype('float16')
all_prior_orders['prices']=all_prior_orders['prices'].astype('float16')
all_prior_orders['order_frequency_median']=all_prior_orders['order_frequency_median'].astype('float16')
```

```python
#user_id is the shared variable
#must use a different indicator variable since _merge already exists in dataset
customers_orders = customers.merge(all_prior_orders, on = 'user_id', how = 'inner', indicator = 'big_merge')
```

```python
#let's double check the merge with the indicator value
customers_orders['big_merge'].value_counts()
```
